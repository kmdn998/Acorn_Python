### 2) 문자열 해결을 위한 데이터 인코딩
- 레이블 인코딩
- 원 핫 인코딩

#### 레이블 인코딩: Category Feature를 코드형 숫자값으로 변환

```python
from sklearn.preprocessing import LabelEncoder

items=['TV','냉장고','전자렌지','컴퓨터','선풍기','믹서','믹서','선풍기']

encoder= LabelEncoder()
encoder.fit(items)
labels= encoder.transform(items)
```

```python
print(labels)
print(encoder.classes_)
print(encoder.inverse_transform([0,1,2,3]))
```

#### 원 핫 인코딩
- 새로운 feature를 추가해 고유 갑세 해당하는 컬럼에만 1을 표시하고 나머지는 0으로 표시
- 사전에 반드시 숫자로 변환이 되어있어야 한다
- 입력값으로 2차원 데이터가 필요하다

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
items=['TV','냉장고','전자렌지','컴퓨터','선풍기','믹서','믹서','선풍기']

encoder= LabelEncoder()
encoder.fit(items)
labels= encoder.transform(items)

labels= labels.reshape(-1,1) # 원핫인코딩은 2차원이여야함

onehot= OneHotEncoder()
onehot.fit(labels)
oh= onehot.transform(labels)
```

```python
oh.toarray()
```

### 3) Feature Scaling
- 표준화: (Xi - mean(X)) / sd(X)
- 정규화: (Xi - min(X)) / (max(X) - min(X))

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

#### StandardScaler 표준화

```python
from sklearn.datasets import load_iris

iris= load_iris()
df= pd.DataFrame(data= iris.data, columns= iris.feature_names)
df
```

```python
print('feature들의 평균:\n', df.mean())
print('----------------')
print('feature들의 분산:\n', df.var())
```

```python
scaler= StandardScaler()
scaler.fit(df)
iris_scaled=scaler.transform(df)
iris_scaled # 데이터 표준화 변환
```

```python
pd.DataFrame(data=iris_scaled, columns=iris.feature_names)
print('feature들의 평균:\n', df.mean())
print('----------------')
print('feature들의 분산:\n', df.var())
```

#### MinMaxScaler 정규화

```python
scaler= MinMaxScaler()
scaler.fit(df)
iris_scaled=scaler.transform(df)
iris_scaled # 데이터 정규화 변환
```

```python
pd.DataFrame(data=iris_scaled, columns=iris.feature_names)
print('feature들의 평균:\n', df.mean())
print('----------------')
print('feature들의 분산:\n', df.var())
```

### Scaling을 할 때 주의할 점
같은 부류의 데이터는 fitting 한번만

```python
train= np.arange(0,11).reshape(-1,1)
test= np.arange(0,6).reshape(-1,1)
print(train)
print(test)
```

```python
scaler= MinMaxScaler()
scaler.fit(train)
train_scaled= scaler.transform(train)

print('원본:', train.flatten())
print('스케일링: ', train_scaled.flatten())

print('-----------------------------------')

test_scaled= scaler.transform(test)

print('원본:', test.flatten())
print('스케일링: ', test_scaled.flatten())
```

## (1) 통계적 방법

```python
import statsmodels.api as sm
```

```python
# 독립변수와 종속변수 나누기
# X= df.loc[0:11] 모델 안돌아감
X= df[df.columns[0:11]]
y= df[['quality']]
```

```python
# 선형 회귀 분석. 
# 최소 제곱법: 오차를 뚜렷하게 하기위해 제곱

X_add= sm.add_constant(X) # y절편 const 추가

# y절편 없음. 기울기만 존재
multi_model= sm.OLS(y, X_add).fit()
print(multi_model.summary())
# R-squared. 설명력: 0.35
# Prob. p-value: 독립변수간 차이가 존재한다
```

## (2) 머신러닝 방법: 선형 회귀

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# feature(특성, 훈련 데이터)와 label(target, 정답 데이터) 나누기
X= df[df.columns[0:11]]
y= df[['quality']]
```

```python
# 훈련 데이터와 테스트 데이터 나누기
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=.3, random_state=42)
```

```python
# 훈련(학습)
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
# 평가
print(model.score(X_train, y_train)) # 설명력
print(model.score(X_test, y_test))
```

```python
# 예측(서비스 런칭)
pred= model.predict(X_test) # 새로운 데이터로 실행
pred
```
