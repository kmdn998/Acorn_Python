## (5) 영화 분류 프로그램
knn: 이웃과의 거리로 분류

```python
import operator

def createDataset():
    group=[[1,2], [1,4], [4,1], [4,2]] # [kiss, kick]
    label= ['Action','Action','Romance','Romance']
    return group, label

def classify(X_new, X, y, k):
    # 거리 측정
    diffMat= np.array(X_new) -np.array(X) # array로 변환
#     print(diffMat)
    
    # L2 norm
    sqDiffMat= diffMat**2
    distance= (sqDiffMat.sum(axis=1))**0.5
#     print(distance)
    
    # 이웃의 갯수로 묶어서 클래스로 분류
    classCount= {}
    
    sort_distance= distance.argsort() # 인덱스로 정렬
    print(sort_distance)
    for i in range(k):
        voteLabel= y[sort_distance[i]]
        classCount[voteLabel]= classCount.get(voteLabel,0) +1
#     print(classCount)

    return sorted(classCount.items(), key=operator.itemgetter(1), reverse=True)[0][0]
    # itemgetter(0)은 아스키코드순으로 정렬
```

```python
X, y= createDataset()
result= classify([2,3], X, y,3)
print('결과는 {}이다.'.format(result))
```

# 4.검증(Validation): Cross Validation
훈련데이터로도 자체test 진행
- K-fold 방식

성능이 좋아지진 않음. 과적합 방지용

```python
from sklearn.model_selection import KFold
```

```python
iris=load_iris()
X_train, X_test, y_train, y_test= train_test_split(iris.data, iris.target, random_state=0)
```

```python
kf = KFold(n_splits=10) # 5개의 군집으로 나누기
list(kf.split(X_train))
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
cv_acc=[]

for train_index, test_index in kf.split(X_train):
    X_train1, X_test1= X_train[train_index], X_train[test_index]
    y_train1, y_test1= y_train[train_index], y_train[test_index]
#     print(X_train1.shape, X_test1.shape)

    knn.fit(X_train1, y_train1)
    
    pred= knn.predict(X_test1)
    
#     cv_acc.append(knn.score(X_test1, y_test1))
    cv_acc.append(accuracy_score(y_test1, pred)) # 모델명 쓸 필요 없음
```

```python
print(cv_acc)
print(np.mean(cv_acc))
```

```python
print(knn.score(X_test, y_test))

pred= knn.predict(X_test)
print(accuracy_score(y_test, pred))
```

파라미터를 직접 지정하느냐, 랜덤으로 지정하느냐 차이

```python
iris=load_iris()

X_train, X_test, y_train, y_test= train_test_split(iris.data, iris.target, random_state=0)

knn= KNeighborsClassifier(n_neighbors=3)
```

```python
?GridSearchCV
```

```python
params={'n_neighbors':[1,3,5,7,9,11]}

grid_knn= GridSearchCV(knn, params, cv=5) # 객체 생성
grid_knn.fit(X_train, y_train)
```
