# 2. 알고리즘 비교
- L2 Norm 규제(Regularization) 사용
- C 파라미터를 사용해서 규제의 강도를 설정

```python
X, y= mglearn.datasets.make_forge()

_, axes= plt.subplots(1, 2, figsize=(10,3))
for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    clf= model.fit(X, y)
    
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    mglearn.plots.plot_2d_separator(clf, X, ax=ax)
    
    ax.set_title(clf.__class__.__name__)
```

# 3. C값에 따른 변화

```python
cancer= load_breast_cancer()
cancer.data.shape
```

```python
X_train, X_test, y_train, y_test= train_test_split(cancer.data, cancer.target, random_state=0)
```

```python
logreg= LogisticRegression(C=1, max_iter=10000).fit(X_train, y_train)

print('훈련 점수: ', logreg.score(X_train, y_train))
print('테스트 점수: ', logreg.score(X_test, y_test))
```

```python
logreg1000= LogisticRegression(C=1000, max_iter=10000).fit(X_train, y_train)

print('훈련 점수: ', logreg.score(X_train, y_train))
print('테스트 점수: ', logreg.score(X_test, y_test))
# C가 커지니까 과적합
# C가 클수록 모델이 복잡. 규제 약해짐
```

```python
logreg0001= LogisticRegression(C=0.001, max_iter=10000).fit(X_train, y_train)

print('훈련 점수: ', logreg.score(X_train, y_train))
print('테스트 점수: ', logreg.score(X_test, y_test))
```

```python
plt.plot(logreg.coef_.T, '^', label='C=1')
plt.plot(logreg1000.coef_.T, 's', label='C=1000')
plt.plot(logreg0001.coef_.T, 'v', label='C=0.001')

plt.legend()
plt.xlabel('계수 목록')
plt.ylabel('계수 크기')
plt.hlines(0,0,cancer.data.shape[1])
plt.ylim(-6,6)
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
```

```python
logreg_l1= LogisticRegression(C=1, max_iter=10000, penalty='l1', solver='saga').fit(X_train, y_train)

print('훈련 점수: ', logreg_l1.score(X_train, y_train))
print('테스트 점수: ', logreg_l1.score(X_test, y_test))

logreg1000_l1= LogisticRegression(C=1000, max_iter=10000, penalty='l1', solver='saga').fit(X_train, y_train)

print('훈련 점수: ', logreg_l1.score(X_train, y_train))
print('테스트 점수: ', logreg_l1.score(X_test, y_test))

logreg0001_l1= LogisticRegression(C=0.001, max_iter=10000, penalty='l1', solver='saga').fit(X_train, y_train)

print('훈련 점수: ', logreg_l1.score(X_train, y_train))
print('테스트 점수: ', logreg_l1.score(X_test, y_test))
```

## (1) 오차 행렬
- 이진 분류의 예측 오류가 얼마인지와 더부어 어떠한 유형의 예측 오류가 발생하고 있는지를 함께 나타내는 지표
- TN, TP, FN, FP

```python
from sklearn. datasets import load_digits
# 숫자 중 7을 분류하기
```

```python
digits= load_digits()
print(digits.data.shape)
print(digits.data)
print(digits.target)
```

```python
y= (digits.target ==7).astype(int)
```

```python
pd.Series(y).value_counts()
```

```python
plt.imshow(digits.data[-1].reshape(8,8))
```

```python
from sklearn.base import BaseEstimator
class MyFakeClassifier(BaseEstimator):
    def fit(self, x, y):
        pass
    
    def predict(self, X):
        return np.zeros((len(X),1), dtype=bool) # 무조건 7이 아니라고 예측
    
```

```python
X_train, X_test, y_train, y_test= train_test_split(digits.data, y, random_state=42)

model= MyFakeClassifier()
model.fit(X_train, y_train)

pred= model.predict(X_test)
print('정확도: ', accuracy_score(y_test, pred))
```

```python
confusion_matrix(y_test, pred) # (TN, FN, FP, TP)
```

## (2) 정확도
accuracy_score
- 예측 결과가 동일한 데이터 건수 / 전체 예측 데이터 건수
- (TP + TN) / (TN + TN + FP + FN)

```python
( 409 + 0) / ( 409 + 0 + 41 + 0)
```

## (3) 정밀도와 재현율

 - Positive에 예측 성능을 좀 더 초점 맞춘 평가 지표
 
 
 - 정밀도: TP / (FP + TP)
     - 예측을 Positive로 한 대상 중에 예측과 실제값이 Positive로 일치한 데이터의 비율
     - 실제 음성 데이터를 양성으로 잘못 판단하게 되면 업무상 큰 영향이 발생하는 경우에 중요한 지표
         - 스팸메일판정
         
         
 - 재현율: TP / (FN + TP)
    - 실제값이 Positive인 대상 중에 예측과 실제값이 Positive로 일치한 데이터의 비율
    - 실제 양성 데이터를 음성으로 잘못 판단하게되면 업무상 큰 영향이 발생하는 경우에 중요한 지표
        - 암 환자 판정, 보험 사기 판정
    - 민감도(Sensitive) 또는 TPR(True Positive Pate)라고도 한다. 

## (4) F1 Score

- 정밀도와 재현율의 조화평균(Harmonic Mean)
- ((정밀도 * 재현율) / ( 정밀도 + 재현율)) * 2

## (5) ROC, AUC

```python
from sklearn.datasets import make_classification
```

```python
from sklearn.datasets import make_classification

X, y= make_classification(n_samples=1000, n_features=2, random_state=0, n_informative=2, n_redundant=0)

mglearn.discrete_scatter(X[:,0], X[:,1], y)
```

```python
model= LogisticRegression().fit(X, y)
```

```python
y_hat= model.predict(X)
```

```python
print(y_hat[:30])
print(y[:30])
```

```python
confusion_matrix(y, y_hat)
```

```python
print(classification_report(y, y_hat))
```

```python
fpr, tpr, thresholds= roc_curve(y, model.decision_function(X))
# fpr: False Positive Rate 양성으로 잘못 예측한 경우
# tpr: 재현율
```

```python
# plt.plot([0,1], [0,1], 'k--')
plt.plot(fpr, tpr, 'o--')
# 커브선이 왼쪽 상단에 가까울수록 좋은 성능
# 이진분류에서만 사용가능
```
