## (3) 검증과 튜닝을 동시에
- 교차 검증과 최적의 하이퍼 파라미터 튜닝을 한번에 처리
- GridSearchCV()
- RandomSearchCV()

파라미터를 직접 지정하느냐, 랜덤으로 지정하느냐 차이

```python
iris=load_iris()

X_train, X_test, y_train, y_test= train_test_split(iris.data, iris.target, random_state=0)

knn= KNeighborsClassifier(n_neighbors=3)
```

### 2) RandomSearchCV()

```python
from scipy.stats import randint, uniform
params={'n_neighbors': randint(low=1, high=9)}

random_knn=RandomizedSearchCV(knn, params, cv=5, n_iter=10)
random_knn.fit(X_train, y_train)
```

```python
pred= random_knn.predict(X_test)
accuracy_score(y_test, pred)
```

## (1) 예제1
- 1 -> 1, ... 100 -> 100. 회귀문제
- wx + b: w=1, b=0

```python
import tensorflow as tf
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD, Adam
```

```python
X_train = [1, 2, 3, 4, 5]
y_train = [2.1, 3.1, 4.1, 5.1, 6.1]

model= tf.keras.models.Sequential() 

model.add(Flatten(input_shape=(1,))) 

model.add(Dense(1, activation='linear')) 

model.compile(loss='mse', optimizer=SGD()) 
model.summary()

model.fit(X_train, y_train, epochs=200) 
```
